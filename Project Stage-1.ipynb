```python
import urllib.parse,urllib.request,urllib.error
import pandas as pd

#!conda install -c conda-forge beautifulsoup4 --yes
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

html = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(html,"html.parser")
```

```python
# Then inserting the table data of the wikipedia page 
table = soup.find_all('table')[0] 

# Now loading the Data to list using pandas read_html
lst = pd.read_html(str(table))

# Converting List to pandas data frame 
df = pd.DataFrame(lst[0])
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    nearby_venues = []
    LIMIT = 100
    for name, lat, lng in zip(names, latitudes, longitudes):
        #print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            str(v['venue']['name']), 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['id'],
            str(v['venue']['categories'][0]['name'])) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category id',
                  'Venue Category']
    nearby_venues.head()
    
    return(nearby_venues)
```

```python
Toronto_onehot = pd.get_dummies(Toronto_nearby["Venue Category"])

#adding the neighbourhood column
Toronto_onehot["Neighborhood"] = Toronto_nearby["Neighborhood"]

# Moving the Neighborhood olumn to first 
column_sequence = ["Neighborhood"] + list(Toronto_onehot.columns[:159]) + list(Toronto_onehot.columns[161:]) 
print(Toronto_onehot.columns.get_loc("Neighborhood"),Toronto_onehot.columns.get_loc("Yoga Studio"))

Toronto_onehot = Toronto_onehot[column_sequence] 

Toronto_onehot.head()
```

```python
Toronto_groupby = Toronto_onehot.groupby("Neighborhood").mean().reset_index()
Toronto_groupby
```

```python
num_top_venues = 5

for hood in Toronto_groupby['Neighborhood']:
    print("----"+hood+"----")
    temp = Toronto_groupby[Toronto_groupby['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
import numpy as np
column_names = ["Neighborhood"]

for i in range(1,11):
    column_names.append(str(i) + " Most Visited Place")

neighborhoods_venues_sorted = pd.DataFrame(columns=column_names)
neighborhoods_venues_sorted['Neighborhood'] = Toronto_groupby['Neighborhood']

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

for ind in np.arange(Toronto_groupby.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Toronto_groupby.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```
